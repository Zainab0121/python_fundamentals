##### THE GUESSING GAME

```python
secret_code = 37
guess_count = 0
guess_limit = 3
while guess_count < guess_limit:
    guess = int(input('What is the secret code? '))
    guess_count +=1
    if guess < secret_code:
        print ('Go higher')
    elif guess > secret_code:
        print ('Go lower')
    elif guess == secret_code:
        print('Genius! You cracked the code')
        break
else:
    print ('Better luck next time')
```

##### CAR GAME

```python
command = ""
started = False
while True: # The loop is constantly executed until explicitly broken
    command = input('>>>').lower()
    if command == 'start':
        print("Let's go.......VROOOM")
        if started:
            print('We are already on the move')
        else:
            started = True
    elif command == 'stop':
              if not started:
                  print('Already stopped')
              else:
                  started = False
                  print ('HALTTTT!!!!')
    elif command == 'help':
        print(''' 
     HERE IS YOUR GUIDE:-
        1. Start - Starts the car
        2. Stop - Stops the car
        3. Quit- Quits the game ''')
    elif command == 'quit':
        print('Sorry to see you leave.... Come back soon')
        break # stops the entire loop immediately   
    else:
        print('Invalid Input')
```

### FUNCTIONS
- Block of code that performs a specific task

```python
def greetings():
    print('Hello') # only executed if the function is called
    print('Habari Yako')

# always add two line breaks after defining a function
print('Start') # execution starts here
greetings() # this calls the function
print('Finish')
```

- PARAMETERS:- placeholders for receiving information
- ARGUMENTS:- actual value we supply to a function
- - positional arguments:- position based on the parameters matters
  - keyword arguments:- position based on the parameters does not matter

```python
def greetings(firstname, lastname): # These are parameters
    print(f'Hello, {firstname} {lastname}')
    print('How you doin 😎')


print('Start') 
greetings('Joey', 'Tribbiani')# These are arguments
print('Finish')
```

```python
def billing(customer_name, items, actual_price, shipping_cost, discount):
    print(f'''
        Hello, {customer_name}!
 Your billing is as follows:-
 Items: {items}
 Price: {actual_price} 
 Shipping: {shipping_cost} 
 Discount: {discount}''')

billing('Penny', 'Shoes', actual_price = 2500, shipping_cost = 150, discount = 20)
# Here we have both positional and keyword arguments.
# positional arguments are the name and the item
# keyword arguments are the pricings
# Note that positional arguments must come before keyword arguments
```

```python
def square(num):
    return num * num # return value
# by default all functions in python return None unless return value is defined
result = square(int(input('What is the square of: ')))
print(result)
```

```python
def emoji_converter(message):
    words = message.split(' ') 
    emojis = {
    ':)' : '😁',
    ':(' : '😔'}
    output = ''
    for word in words:
        output += emojis.get(word, word) + ' '
    return output
    
message = input('>>>')
results = emoji_converter(message)
print (results)    
```

##### INHERITANCE

```python
class Mammal:
    def walk(self):
        print('Crawl, Walk, Run: JUST MOVE')

        
class Dog(Mammal):
    pass # A class is not permitted to be empty

class Cat(Mammal):
    def sound(self):
        print("Meow")

dog1 = Dog()
dog1.walk()

cat1 = Cat()
cat1.walk()
cat1.sound()
```

### MODULES 
- IMPORTING RANDOM MODULE TO GENERATE RANDOM VALUES

```python
import random

class Dice:
    def roll(self):
        first = random.randint(1,6)
        second = random.randint(1,6)
        return first, second

dice = Dice()
print (dice.roll())
```
